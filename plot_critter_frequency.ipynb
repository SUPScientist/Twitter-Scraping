```python
import pandas as pd
from collections import Counter
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df_retwe = pd.read_csv('tweet_replies.csv')
df_retwe.head()
```

```python
all_the_words_nest = []

for index, row in df_retwe.iterrows():
    row_words = row.text.split(" ")
    all_the_words_nest.append(row_words)
    
all_the_words = [item for sublist in all_the_words_nest for item in sublist]

all_the_words
```

```python
# counts = Counter(all_the_words)

n = 25
counts = dict(Counter(all_the_words).most_common(n))

labels, values = zip(*counts.items())
```

```python
# sort your values in descending order
indSort = np.argsort(values)[::-1]

# rearrange your data
labels = np.array(labels)[indSort]
values = np.array(values)[indSort]

indexes = np.arange(len(labels))

bar_width = 0.35

plt.figure(figsize = (10, 10))
plt.bar(indexes, values)

# add labels
plt.xticks(indexes + bar_width, labels)
plt.show()
```
